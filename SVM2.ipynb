```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.inspection import DecisionBoundaryDisplay
```

```python
# create a non-linear i.e Circular Decision boundary dataset
from sklearn.datasets import make_circles
X, y = make_circles(100, factor=.1, noise=.1, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

### Fit SVC: Linear and RBF
SVC Doc: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn-svm-svc

```python
from sklearn.svm import SVC

# try using svm with linear kernel
linear_clf = SVC(kernel='linear')
linear_clf.fit(X, y)
```

```python
# plot the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, s=50)
plt.title("SVM with linear kernel on non-linear dataset")

ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    linear_clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    linear_clf.support_vectors_[:, 0],
    linear_clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```

```python
### rbf
rbf_clf = SVC(kernel='rbf', C=1E6)
rbf_clf.fit(X, y)
```

```python
# plot the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, s=50)
plt.title("SVM with linear kernel on non-linear dataset")

ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    rbf_clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    rbf_clf.support_vectors_[:, 0],
    rbf_clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
